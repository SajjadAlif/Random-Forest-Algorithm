```python
#Split the dataset into X and y
X=df.iloc[5:,[1]].values
y=df['Salary'].values
print(X.shape)
print(y.shape)
```

```python
#Apply Linear Regression and check whether the regession line fits all the observations or not

from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(X,y)
y_predict=model_lr.predict(X)

plt.scatter(X,y,color='red')
plt.plot(X,y_predict,color="blue")
plt.title("Linear Regression Model (Level Vs Salary)")
plt.xlabel("Employee's Level")
plt.ylabel("Salary")
plt.show()
```

From the above it is clearly shown that the regression line doesn't fit all the observations.

```python
#Apply the polynomial regression on this same dataset and see how this is working.
#Apply degree=2

from sklearn.preprocessing import PolynomialFeatures
polynomial=PolynomialFeatures(degree=2)
X_poly=polynomial.fit_transform(X)
model_plr=LinearRegression()
model_plr.fit(X_poly,y)
y_predict_poly=model_plr.predict(X_poly)

plt.scatter(X,y,color="green")
plt.plot(X,y_predict_poly,color="red")
plt.title("Polynomial Regression (Position Level Vs Salary)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
#Apply Degree = 3

from sklearn.preprocessing import PolynomialFeatures
polynomial=PolynomialFeatures(degree=3)
X_poly=polynomial.fit_transform(X)
model_plr=LinearRegression()
model_plr.fit(X_poly,y)
y_predict_poly=model_plr.predict(X_poly)

plt.scatter(X,y,color="green")
plt.plot(X,y_predict_poly,color="red")
plt.title("Polynomial Regression (Position Level Vs Salary)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
#Apply Degree = 4

from sklearn.preprocessing import PolynomialFeatures
polynomial=PolynomialFeatures(degree=4)
X_poly=polynomial.fit_transform(X)
model_plr=LinearRegression()
model_plr.fit(X_poly,y)
y_predict_poly=model_plr.predict(X_poly)

plt.scatter(X,y,color="green")
plt.plot(X,y_predict_poly,color="red")
plt.title("Polynomial Regression (Position Level Vs Salary)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

From the above graph we can see that on degree : 4, the curve covers all the observations.



```python
#Predict the model with new or unseen data
#First try with Linear regression
list_level=[3,12,11.5,6.5,7,8,9,3,10.5]
list_level=np.reshape(np.array(list_level),newshape=(len(list_level),1))
print("Prediction based on Linear Regression")
print(model_lr.predict(list_level))
print("\n")
#First try with Ploynomial Linear regression
list_level=[3,12,11.5,6.5,7,8,9,3,10.5]
list_level=np.reshape(np.array(list_level),newshape=(len(list_level),1))
list_level=polynomial.transform(list_level)
print("Prediction Based on Polynomial Regression")
print(model_plr.predict(list_level))
```

From this prediction we can tell that poly regression algorithm works fine compare to linear regression
